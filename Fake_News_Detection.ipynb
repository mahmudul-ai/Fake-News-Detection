```python
def clean_text(text):
    # Remove HTML tags
    text = re.sub('<.*?>', '', text)

    # Remove special characters and numbers
    text = re.sub('[^a-zA-Z\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove stopwords and lemmatize
    words = [lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words]

    # Remove redundant whitespaces
    text = ' '.join(words)

    return text

def undersample(X, y):
    # Identify the majority and minority classes
    class_counts = Counter(y)
    majority_class = max(class_counts, key=class_counts.get)
    minority_class_count = min(class_counts.values())

    # Undersample the majority class
    X_undersampled = []
    y_undersampled = []
    for label in set(y):
      X_class = X[y == label]
      if label == majority_class:
        indices = np.random.choice(X_class.index, minority_class_count, replace=False)
        X_undersampled.extend(X_class.loc[indices].values.tolist())
        y_undersampled.extend(y[indices].tolist())
      else:
        X_undersampled.extend(X_class.values.tolist())
        y_undersampled.extend(y[X_class.index].tolist())

    return pd.DataFrame(X_undersampled, columns=X.columns), pd.Series(y_undersampled)
```

```python
import nltk
from nltk import ngrams
from collections import Counter

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def analyze_text(text, clean = True, inplace=False):
    # Count exclamation marks
    exclamation_count = text.count('!')

    if clean:
      text = clean_text(text)

    # Tokenize the text
    tokens = nltk.word_tokenize(text)
    # Perform N-gram analysis (e.g., trigrams)
    n = 3  # Adjust 'n' for different N-grams
    trigrams = list(ngrams(tokens, n))
    trigram_counts = Counter(trigrams)

    # Identify frequent N-grams (example: top 5)
    frequent_trigrams = trigram_counts.most_common(5)

    # Calculate lexical diversity (Type-Token Ratio)
    types = set(tokens)
    ttr = len(types) / len(tokens) if tokens else 0

    # Perform part-of-speech tagging
    pos_tags = nltk.pos_tag(tokens)

    # Example: Find adjective-heavy structures (adjust as needed)
    adjective_count = sum(1 for word, tag in pos_tags if tag.startswith('JJ'))

    out = {
          "frequent_trigrams": frequent_trigrams,
          "ttr": ttr,
          "exclamation_count": exclamation_count,
          "adjective_count": adjective_count,
      }
    if inplace:
      out['statement-clean'] = text
      # print(text)
    return out

# Apply the analysis to the 'statement' column and assign the results to new columns
out_cols = ['statement-clean', 'frequent_trigrams', 'ttr', 'exclamation_count', 'adjective_count']
out = X_train_undersampled['statement'].apply(analyze_text, inplace=True).apply(pd.Series)
for col in out_cols:
  X_train_undersampled[col] = out[col]

out = X_test['statement'].apply(analyze_text, inplace=True).apply(pd.Series)
for col in out_cols:
  X_test[col] = out[col]

out = X_val['statement'].apply(analyze_text, inplace=True).apply(pd.Series)
for col in out_cols:
  X_val[col] = out[col]


# X_test[out_cols] = X_test['statement'].apply(analyze_text, inplace=True).apply(pd.Series)
# X_val[out_cols] = X_val['statement'].apply(analyze_text, inplace=True).apply(pd.Series)

# Display the results
print(X_train_undersampled.columns)
X_train_undersampled[out_cols].head()
```

```python
import nltk
import spacy
import torch
from itertools import combinations
from transformers import pipeline

# Load required NLP tools
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

nlp = spacy.load('en_core_web_sm')  # Load spaCy's English model

# Check if GPU is available
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")


# Load the RoBERTa model for Natural Language Inference (NLI)
device = "cuda" if torch.cuda.is_available() else "cpu"
nli_classifier = pipeline("text-classification", model="roberta-large-mnli", device=0 if device == "cuda" else -1)

def extract_clauses(sentence):
    """
    Extracts clauses from a sentence using POS tagging and dependency parsing.

    Args:
    - sentence (str): The input sentence.

    Returns:
    - list: Extracted clauses.
    """
    doc = nlp(sentence.lower())
    # text = ' '.join(words)
    clauses = []
    clause = []

    for token in doc:
        clause.append(token.text)

        # Clause boundary indicators: punctuations (commas, semicolons) and conjunctions
        if token.dep_ in {"cc", "mark", "punct", "relcl"} or token.text in {",", ";", ".", "and", "but", "because", "although", "if", "when", "while", "hence"}:
            if clause:
                clauses.append(" ".join(clause).strip())
                clause = []

    # Add any remaining clause
    if clause:
        clauses.append(" ".join(clause).strip())

    return clauses

def detect_contradictions(statement):
  """
  Detect contradiction between two sentences using RoBERTa-large-MNLI.

  Args:
  - premise (str): The context sentence.
  - hypothesis (str): The reply sentence.

  Returns:
  - dict: Contains label ('ENTAILMENT', 'NEUTRAL', or 'CONTRADICTION') and confidence score.
  """
  clauses = extract_clauses(statement)
#   print(len(clauses))

  # Generate all possible clause pairs
  clause_pairs = list(combinations(clauses, 2))

  contradiction_count = 0
  total_pairs = len(clause_pairs)

  for premise, hypothesis in clause_pairs:
    # print(f"🧐 Premise: {premise} | Hypothesis: {hypothesis}")
    result = nli_classifier(f"{premise} </s> {hypothesis}")
    label_scores = {res["label"]: res["score"] for res in result}
    contradiction_score = label_scores.get("CONTRADICTION", 0.0)

    out = {
        "premise": premise,
        "hypothesis": hypothesis,
        "label": max(label_scores, key=label_scores.get),
        "contradiction_score": contradiction_score
    }
    # print(out["label"])

    if out["label"].lower() == 'contradiction':
        contradiction_count += 1

  contradiction_score = contradiction_count / total_pairs if total_pairs > 0 else 0

  return {"contradiction_score": contradiction_score}

```

```python
X_train_undersampled["contradiction_score"] = X_train_undersampled["statement"].apply(detect_contradictions).apply(pd.Series)
# # Example sentence
# test_sentence = "He is a known theif hence people trust him"
# result = detect_contradictions(test_sentence)
# print("\nFinal Contradiction Score:", result)

X_train_undersampled[["statement", "contradiction_score"]].head()
```
